```python
import pandas as pd

#Load the data and display 10 rows:

df = pd.read_excel('Online Retail.xlsx') 
print(df.head(10))
```

```python
#Check rows and missing values

print(df.shape)
print(df.isnull().sum())
```

```python
# Remove duplicate values

print('Duplicates before removing:',df.duplicated().sum()) 
df = df.drop_duplicates() 
print('Duplicates after removing:', df.duplicated().sum())
```

```python
# Add Revenue column

df['Revenue'] = df['Quantity'] * df['UnitPrice'] 
print(df[['Quantity', 'UnitPrice', 'Revenue']].head())
```

```python
# Top 10 selling products based on quantity

top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Calculate total revenue of the store

total_revenue = df['Revenue'].sum() 
print(f'Total Revenue = {total_revenue:.2f}')
```

```python
# Total number of unique customers

unique_customers = df['CustomerID'].nunique() 
print(f'Unique Customers = {unique_customers}')
```

```python
# Calculate average order value

avg_order_value = df.groupby('InvoiceNo')['Revenue'].sum().mean() 
print(f'Average Order Value = {avg_order_value:.2f}')
```

```python
# Country with highest number of orders

df.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False).head(1)
```

```python
# Products bought together

basket = df.groupby('InvoiceNo')['Description'].apply(lambda x: ', '.join(x.astype(str)))
print(basket.head(10))
```

```python
# Customers who purchased more than 10 items

customers_10 = df.groupby('CustomerID')['Quantity'].sum() 
customers_10 = customers_10[customers_10 > 10] 
print(customers_10.head(10))
```

```python
# Peak shopping hours

df['Hour'] = pd.to_datetime(df['InvoiceDate']).dt.hour
peak_hours = df.groupby('Hour')['InvoiceNo'].count().sort_values(ascending=False)
print(peak_hours.head())
```
